# DATA UNDERSTANDING

## Important parameters:

- Number of instances: 1025
- Number of attributes: 14 continuous attributes

## Each of the attributes:

The variables in the heart disease dataset are as follows:

1. **`Age`:** The age of the patient, typically recorded in years.
2. **`Gender`:** The gender of the patient, often encoded numerically (0 for female, 1 for male).
3. **`ChestPain`:** The type of chest pain experienced by the patient, categorized into different types, such as typical angina, atypical angina, non-anginal pain, or asymptomatic.
4. **`RestingBP` (Resting Blood Pressure):** The blood pressure of the patient while at rest, usually measured in millimeters of mercury (mm Hg).
5. **`Cholesterol`:** The serum cholesterol level of the patient, measured in milligrams per deciliter (mg/dL).
6. **`FastingBS` (Fasting Blood Sugar):** A categorical variable indicating whether the patient's fasting blood sugar is greater than 120 mg/dL (1 for true, 0 for false).
7. **`RestECG` (Resting Electrocardiographic Results):** The results of the resting electrocardiogram, which may show normal results or indicate the presence of ST-T wave abnormalities or left ventricular hypertrophy.
8. **`MaxHR` (Maximum Heart Rate):** The maximum heart rate achieved by the patient during exercise.
9. **`ExerciseAngina`:** A categorical variable indicating whether the patient experienced exercise-induced angina (1 for yes, 0 for no).
10. **`STDepression` (ST Depression):** The ST segment depression induced by exercise relative to rest.
11. **`STSlope` (Slope of the Peak Exercise ST Segment):** The slope of the peak exercise ST segment, categorized as upsloping, flat, or downsloping.
12. **`NumVessels` (Number of Major Vessels):** The number of major blood vessels colored by fluoroscopy, usually ranging from 0 to 3.
13. **`Thalassemia`:** A categorical variable indicating the presence of thalassemia, classified as normal, fixed defect, or reversible defect.
14. **`HeartDisease`:** The target variable indicating the presence or absence of heart disease, often encoded numerically (0 for no disease, 1 for presence of disease).

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Identify and categorize highly correlated variable pairs
highly_correlated_pairs = []

# Iterate through column names
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        correlation_value = correlation_matrix.iloc[i, j]
        
        # Determine the strength of the relationship
        if abs(correlation_value) < 0.25:
            relationship_strength = "No relationship"
        elif 0.25 <= abs(correlation_value) < 0.5:
            relationship_strength = "Weak relationship"
        elif 0.5 <= abs(correlation_value) < 0.75:
            relationship_strength = "Moderate relationship"
        else:
            relationship_strength = "Strong relationship"
        
        # Add the pair of column names and the strength of the relationship to the list
        pair = (correlation_matrix.columns[i], correlation_matrix.columns[j], relationship_strength)
        highly_correlated_pairs.append(pair)

# Print the highly correlated variable pairs with relationship strength
for pair in highly_correlated_pairs:
    print(f"{pair[0]} and {pair[1]} | Strength: {pair[2]}")
```
